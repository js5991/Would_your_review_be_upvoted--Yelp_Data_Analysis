###  Working List
1. remove meaningless/duplicate columns such as open/close hour every day,type and name (in list 'Label')
2. remove empty columns
3. rename variable with missing column name
4. change True/False into 1/0 variable
5. recode smoking, alcohol, noise level, attire and wifi variable to categorical variable #check bubsiness this time  
6. get the ambience score for all business

```python
import pandas as pd
business=pd.read_csv('business_restaurant.csv',encoding='cp1252')
business=business.drop(business.columns[0],1)
```

```python
pd.set_option('display.max_columns',None)
```

```python
business=business.dropna(axis=1,how='all')
```

```python
labels =['name','ype','hours.Sunday.open','hours.Wednesday.open','hours.Monday.open','hours.Friday.open','latitude','hours.Monday.close'
,'hours.Sunday.close','hours.Tuesday.close','hours.Thursday.open','full_address','hours.Tuesday.open','neighborhoods','hours.Thursday.close','hours.Saturday.close','hours.Saturday.open'
,'hours.Friday.close','categories','hours.Wednesday.close','longitude','Unnamed: 89','city','BYOB/Corkage']
business=business.drop(labels,1)
```

```python
business.rename(columns={'Unnamed: 14':'Business_star'}, inplace=True)
```

```python
def change_boolean(x):
    if x == False:
        return 0
    elif x==True: 
        return 1
    else:
        return x
business=business.applymap(lambda x: change_boolean(x)) 
```

```python
def edit_alcohol(x):
    if x=='none':
        return 0
    elif (x== 'full_bar' or x=='beer_and_wine'):
        return 1
    else:
        return x
business['Alcohol']=business.Alcohol.map(lambda x: edit_alcohol(x))
```

```python
def change_wifi(x):
    if (x=='paid' or x=='free'):
        return 1
    elif x=='no':
        return 0
    else: return x
business['Wi-Fi']=business['Wi-Fi'].map(lambda x: change_wifi(x))
```

```python
def change_attire(x):
    if (x=='casual'):
        return 1
    elif (x=='dressy' or x=='formal'):
        return 0
    else: return x
business['Attire']=business['Attire'].map(lambda x: change_attire(x))
```

```python
def change_noise(x):
    if (x=='average' or x=='quiet'):
        return 1
    elif (x=='loud' or x=='very_loud'):
        return 0
    else: return x
business['Noise Level']=business['Noise Level'].map(lambda x: change_noise(x))
```

```python
def change_smoke(x):
    if (x=='yes'):
        return 1
    elif (x=='no' or x=='outdoor'):
        return 0
    else: return x
business['Smoking']=business['Smoking'].map(lambda x: change_smoke(x))
```

```python
ambience_label_list=[ i for i in business.columns.values if i.find("Ambience.")>=0]
def change_Ambience(x):
    if x == True:
        return 1
    else: return 0 #or Nan add any number it is nan

for i in ambience_label_list:
    business[i]=business[i].map(lambda x: change_Ambience(x))

business['Ambience_score']=  business[ambience_label_list].sum(axis=1)
business=business.drop(ambience_label_list,1)
```
